### **Verificación de valores faltantes**

Se identifican posibles valores nulos dentro del dataset para determinar si es necesario aplicar algún tratamiento de limpieza.

```python
df.isnull().sum()
```

Se identificó la presencia de valores faltantes en algunas variables numéricas del dataset, principalmente relacionadas con productividad, estrés, sueño y satisfacción laboral.

Debido a que estas variables son relevantes para el análisis y eliminar los registros implicaría una pérdida considerable de información, se decidió reemplazar los valores faltantes utilizando la mediana de cada variable numérica. Esta medida permite conservar la distribución de los datos y reduce el impacto de posibles valores extremos.

```python
df = df.fillna(df.median(numeric_only=True))
df.isnull().sum()
```
